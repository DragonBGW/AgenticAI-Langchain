```python
"""
Minimal Featherless‑AI + LangChain demo
---------------------------------------

Prereqs:
  pip install -U python-dotenv langchain langchain-featherless-ai
  # .env file in the same directory must contain:
  # FEATHERLESSAI_API_KEY=sk-...

Run:  python featherless_demo.py
"""
import getpass
import os
from dotenv import load_dotenv

# 1) Load .env and expose FEATHERLESSAI_API_KEY
load_dotenv()                        # looks for .env in current dir
# (Optional sanity check)
featherless_token = os.getenv("FEATHERLESS_API_KEY")

```

```python
from langchain_featherless_ai import ChatFeatherlessAi

llm = ChatFeatherlessAi(
    model="featherless-ai/Qwerky-72B",
    temperature=0.9,
    max_tokens=None,
    timeout=None,
)

```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate(
    [
        (
            "system",
            "You are a helpful assistant that translates {input_language} to {output_language}.",
        ),
        ("human", "{input}"),
    ]
)

chain = prompt | llm
chain.invoke(
    {
        "input_language": "English",
        "output_language": "German",
        "input": "Tell me what the person in Operation and Management team is telling",
    }
)
```
